```python
#importamos modulos
import cv2
from google.colab.patches import cv2_imshow
```

```python
#cargamos imagen
img = cv2.imread("/content/img9.jpg")
```

```python
#convertimos en escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#cargamos el clasificador de cascada frontal de haar
face_cascade  = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")

#detecta las caras en la imagen
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

#dibujar rectangulos de la cara detectada
for(x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w, y+h), (255,0,0), 2)

#mostar la imagen con las caras detectadas
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
